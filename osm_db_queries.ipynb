```python

# Count of Nodes & Ways

df1 = pd.read_sql_query(
''' 
SELECT COUNT(nodes.id) as Nodes
FROM nodes
''' , conn)

print ('Number of Nodes:')
display(df1)
print  ('\n')

df2 = pd.read_sql_query(
''' 
SELECT COUNT(ways.id) as Ways
FROM ways
''' , conn)

print ('Number of Ways:')
display(df2)
print  ('\n')

```

```python
# Count unique contributing users

df1 = pd.read_sql_query(
''' 
SELECT COUNT(DISTINCT(all_users.uid)) as UserCount          
FROM (SELECT uid FROM nodes UNION ALL SELECT uid FROM ways) all_users
''' , conn)

df1
```

```python
# Count contributions from top 25 contributing users

df = pd.read_sql_query(
''' 
SELECT all_users.user, count(all_users.uid) as contributions
FROM (SELECT uid, user FROM nodes UNION ALL SELECT uid, user FROM ways) as all_users
GROUP BY all_users.uid
ORDER BY Contributions DESC
limit 15;
''' , conn)


df
```

```python
# How many users have only made one contribution?

df = pd.read_sql_query(
''' 
SELECT COUNT(*) as one_contrib_users
FROM
    (SELECT all_users.uid, all_users.user, count(all_users.uid) as contributions
    FROM (SELECT uid, user FROM nodes UNION ALL SELECT uid, user FROM ways) as all_users
    GROUP BY all_users.uid
    HAVING contributions=1) as singles;

''' , conn)


df


```

```python
df = pd.read_sql_query(
''' 
SELECT value, COUNT(*) as num
FROM nodes_tags
WHERE key='amenity'
GROUP BY value
ORDER BY num DESC
LIMIT 35;

''' , conn)


df
```

```python
# List of Pub names


df = pd.read_sql_query(
''' 
SELECT pubs.id, pubs.value, names.value
FROM
    (SELECT id, value
    FROM nodes_tags
    WHERE key='amenity'
    AND value='pub') as pubs 
    JOIN 
    (SELECT id, value
    FROM nodes_tags
    WHERE key='name'
    AND type='regular') as names

ON pubs.id =  names.id
GROUP BY pubs.id

''' , conn)


df
```

```python
#How many distinct pub names?

df = pd.read_sql_query(
''' 
SELECT COUNT(DISTINCT names.value) as num_pub_names
FROM
    (SELECT id, value
    FROM nodes_tags
    WHERE key='amenity'
    AND value='pub') as pubs 
    JOIN 
    (SELECT id, value
    FROM nodes_tags
    WHERE key='name'
    AND type='regular') as names

ON pubs.id =  names.id


''' , conn)


df
```

```python
# Top 10 Pub Names

df = pd.read_sql_query(
''' 
SELECT names.value, COUNT(names.value) as name_count
FROM
    (SELECT id, value
    FROM nodes_tags
    WHERE key='amenity'
    AND value='pub') as pubs 
    JOIN 
    (SELECT id, value
    FROM nodes_tags
    WHERE key='name'
    AND type='regular') as names

ON pubs.id =  names.id
GROUP BY names.value
ORDER BY name_count DESC
LIMIT 10;

''' , conn)


df
```

```python

df = pd.read_sql_query(
''' 
SELECT info.value, count(info.value) as count
FROM
    (SELECT id, value
    FROM nodes_tags
    WHERE key='amenity'
    AND value='cafe') as amenities 
    JOIN 
    (SELECT *
    FROM nodes_tags
    WHERE key!='amenity'
    AND type='regular'
    AND key = 'name') as info

ON amenities.id =  info.id
GROUP BY info.value
ORDER BY count DESC
LIMIT 15;

''' , conn)


df


```
